## 🎨 Stata 画图测试

用 Stata 自带的 `auto.dta`（1978 汽车数据）做几个简单画图，验证 Stata 出图正常（StataMP 18，pystata 后端）。

> 这些是 **Stata** 代码 cell，请用 Stata 运行器 / `stata-code` 运行，而不是上面的 R kernel。

```stata
* 1) 直方图：价格分布
sysuse auto, clear
histogram price, bin(20) normal title("Distribution of Price")
```

```stata
* 2) 散点图：价格 vs 油耗
scatter price mpg, mcolor(navy) title("Price vs MPG")
```

```stata
* 3) 带拟合线的散点图
twoway (scatter price mpg) (lfit price mpg), title("Price vs MPG with Fitted Line")
```

```stata
* 4) 分组箱线图：进口 vs 国产 的价格
graph box price, over(foreign) title("Price by Origin")
```
